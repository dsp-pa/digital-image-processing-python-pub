```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
box = np.zeros((300, 300))
```

```python
print(box)
```

```python
plt.imshow(box)
```

```python
white = np.ones((300, 300)) * 255
print(white)
```

```python
plt.imshow(white, cmap="gray", vmin=0, vmax=255)
```

```python
plt.imshow(box, cmap="gray", vmin=0, vmax=255)
```

```python
white = 255 * np.ones((10, 10))
white[5][5] = 0

plt.imshow(white, cmap="gray", vmin=0, vmax=255)
```

```python
rd = 255 * np.random.random((10, 10))

plt.imshow(rd, cmap="coolwarm", vmin=0, vmax=255)
```

```python
white =  255 * np.ones((10, 10, 3), dtype=np.int8)

white[0][0][1] = 0
white[0][0][2] = 0


white[1][1][0] = 0
white[1][1][2] = 0

white[2][2][0] = 0
white[2][2][1] = 0


white[0][0][0] = 108 
white[0][0][1] = 212
white[0][0][2] = 255


plt.imshow(white)

```
