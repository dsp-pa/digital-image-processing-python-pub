```python
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
from scipy.ndimage import convolve
```

```python
A = Image.open('images/lab 05.png').convert('L')
A = np.asarray(A)
```

```python
plt.imshow(A, cmap='gray', vmin=0, vmax=255)
```

```python
print(A.min())
print(A.max())

plt.hist(A.flatten(), 256);
plt.xlim(0, 255)
```

```python
def normalize_image(g, K=255):
    g = g.astype(np.float64)
    min_g = np.min(g)
    g_m = g - min_g
    max_g_m = np.max(g_m)
    g_s = K * g_m / max_g_m

    return g_s
```

```python
B = normalize_image(A)

Image.fromarray(B.astype(np.uint8)).save('lab05_B.png')
```

```python
plt.imshow(B, cmap="gray")
```

```python
print(B.min())
print(B.max())
# El histograma se puede obtener con el siguiente fragmento de código:

plt.hist(B.flatten(), 256);
```

```python
C = B ** 0.75

C = normalize_image(C)

Image.fromarray(C.astype(np.uint8)).save("lab05_C.png")
plt.figure()
plt.hist(C.flatten(), 256);
```

```python
kernel = [
 [0, 1, 0 ],
 [1, -4, 1],
 [0, 1, 0 ]
]

conv = convolve(C, kernel)
# plt.imshow(conv, cmap="gray")

S = 2
D = C - S * conv
plt.figure()
# plt.imshow(D, cmap="gray")
plt.imshow(D, cmap="gray", vmin=-25, vmax=225)
D = np.clip(D, -25, 225)

D = normalize_image(D)
Image.fromarray(D.astype(np.uint8)).save("lab05_D.png")
```

```python
plt.hist(D.flatten(), 256);
```

```python
D = C - 2 * conv
plt.figure()
plt.imshow(D, cmap="gray")

print(D.max())
plt.imshow(D, cmap="gray", vmin=-10, vmax=225)

D = np.clip(D, -10, 225)
plt.figure()
plt.imshow(D, cmap="gray")

D = normalize_image(D)
Image.fromarray(D.astype(np.uint8)).save("lab05_D.png")
```
