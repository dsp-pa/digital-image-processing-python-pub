```python
box = np.zeros((300, 300))

box[0:74][:] = 1
box[75:149][:] = 0.65
box[150:224][:] = 0.33
plt.imshow(box, cmap="gray")
```

```python
box = np.zeros((300, 300))

for i, x in enumerate(box):
  box[i][:] = i

plt.imshow(box, cmap="gray", vmax=255, vmin=0)

```

```python
box = np.zeros((300, 300))


for i, x in enumerate(box):
  box[i][:] = i

plt.imshow(box, cmap="gray")
```

```python
A = np.zeros((300, 300))

A[0:74][:] = 1
A[75:149][:] = 0.65
A[150:224][:] = 0.33
plt.imshow(A, cmap="gray", vmin=0, vmax=1)


B = np.full((300, 300), 0.5)

C = A + B

plt.imshow(B, cmap="gray", vmin=0, vmax= 1)
plt.imshow(C, cmap="gray", vmin=0, vmax= 1)
```
