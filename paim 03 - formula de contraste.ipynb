```python
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
```

```python
R = Image.open("images/lab03_2023_03_a.jpg").convert("L")
R =  np.asarray(R)
plt.imshow(R, vmin=0, vmax=255, cmap="gray")
```

```python
plt.hist(R.flatten(), bins=50);
```

```python
def normalize_image(g, K=255):
    g = g.astype(np.float64)
    min_g = np.min(g)

    g_m = g - min_g
    max_g_m = np.max(g_m)

    g_s = K * g_m / max_g_m

    return g_s
```

```python
gs = normalize_image(R, 255)
```

```python
plt.hist(gs.flatten(), 50);
```

```python
plt.imshow(gs, vmin=0, vmax=255, cmap="gray")
```
