```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
```

```python
A = Image.open('./images/lab 05.png').convert('L')
A = np.asarray(A)
```

```python
print(A.min())
print(A.max())
```

```python
plt.hist(A.ravel(), 50)
```

```python
plt.hist(A.ravel())
```

```python
plt.hist(A.ravel(), 255)
```

```python
plt.hist(A.flatten(), 50)
```

```python
def normalize_image(g, K=255):
    g = g.astype(np.float64)
    min_g = np.min(g)
    g_m = g - min_g
    max_g_m = np.max(g_m)
    g_s = K * g_m / max_g_m
    return g_s
```

```python
B = normalize_image(A)

Image.fromarray(B.astype(np.uint8)).save('lab05_B.png')
```

```python
plt.hist(B.flatten(), 50)
print(B.min())
print(B.max())
```

```python
C = B ** 0.5
C = normalize_image(C)
plt.imshow(C, cmap="gray")

Image.fromarray(C.astype(np.uint8)).save("lab05_C.png")
plt.figure()

plt.hist(C.flatten(), 50);
```

```python
from scipy.ndimage import convolve

kernel = [
    [0, 1, 0],
    [1, -4, 1],
    [0, 1, 0]
]
conv = convolve(C, kernel)

D = C - 2 * conv

plt.figure()
plt.imshow(D, cmap="gray")

plt.figure()
plt.hist(D.flatten(), 50);

D = D.clip(-20, 225)

plt.figure()
plt.imshow(D, cmap="gray")

D = normalize_image(D)

Image.fromarray(D.astype(np.uint8)).save("lab05_D.png")
plt.figure()

plt.hist(D.flatten(), 50);
```
