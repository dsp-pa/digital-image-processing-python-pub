```python
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
```

```python
A = Image.open('images/lab02.webp').convert("L")
A = np.asarray(A)

print(A.min())
print(A.max())

plt.imshow(A, vmax=255, vmin=0, cmap="gray")
```

```python
r = A
gamma = 2.00
s = A ** gamma

plt.figure(figsize=(6,3), dpi=150)

plt.subplot(121)
plt.imshow(A, cmap="gray")
plt.title("Imagen original")


plt.subplot(122)
plt.imshow(s, cmap="gray")
plt.title("Gamma = 2.00")
plt.savefig("image")
```

```python
r = A
gamma = 0.5
s = A ** gamma

plt.figure(figsize=(6,3), dpi=150)

plt.subplot(121)
plt.imshow(A, cmap="gray")
plt.title("Imagen original")


plt.subplot(122)
plt.imshow(s, cmap="gray")
plt.title(f"Gamma ={gamma}")
plt.savefig("image")
```

```python

plt.figure(figsize=(10,3), dpi=150)

plt.subplot(121)
plt.hist(A.flatten(), bins=50)
plt.title(f"Imagen original")

plt.subplot(122)
plt.hist(s.flatten(), bins=50)
plt.title(f"Imagen transformada")

plt.savefig("image")
```
